```python
def extract_Shows(txt):
    check=False
    data=""
    for i in txt:
        if check==")":
            break
        if i=='(':
        
            check=True
        if check==True:
            data+=1
        return data
        
    
```

```python
df['Shows']=df["listed inside"].apply(extract_Shows)
df
```

```python
df.shape()
```

```python
 # Main Genre (first genre from listed_in)
df['main_genre'] = df['listed inside'].fillna("").apply(lambda x: x.split(",")[0].strip() if x else None)

# 2. Is Recent (1 if released after 2015 else 0)
df['is_recent'] = df['release year'].apply(lambda x: 1 if x >= 2015 else 0)

# 3. Duration in Numbers (movies â†’ minutes, TV shows â†’ seasons)
df['duration_num'] = df['duration'].str.extract(r'(\d+)').astype(float)

# 4. Decade (e.g., 1990s, 2000s)
df['decade'] = (df['release year'] // 10) * 10

# 5. Is Multi-Genre (1 if multiple genres)
df['multi_genre'] = df['listed inside'].apply(lambda x: 1 if (pd.notnull(x) and "," in x) else 0)

# 6. Genre Count (number of genres a show has)
df['genre_count'] = df['listed inside'].fillna("").apply(lambda x: len(x.split(",")) if x else 0)

# 7. Country Count (number of countries listed)
df['country_count'] = df['country'].fillna("").apply(lambda x: len(x.split(",")) if x else 0)

# 8. Region (continent mapping for countries)
continent_map = {
    'India': 'Asia',
    'United States': 'North America',
    'United Kingdom': 'Europe',
    'South Africa': 'Africa',
    'Japan': 'Asia',
    'South Korea': 'Asia',
    'China': 'Asia',
    'Brazil': 'South America',
    'Mexico': 'North America',
    'Canada': 'North America',
    'Australia': 'Oceania',
    'France': 'Europe',
    'Germany': 'Europe',
    'Spain': 'Europe',
    'Italy': 'Europe',
}
df['region'] = df['country'].map(continent_map)

# 9. Content Age Group (kids / family / adult based on listed_in)
def classify_age_group(genres):
    if pd.isnull(genres):
        return "Unknown"
    genres = genres.lower()
    if "kids" in genres or "children" in genres:
        return "Kids"
    elif "family" in genres:
        return "Family"
    else:
        return "Adult"

df['age_group'] = df['listed inside'].apply(classify_age_group)

# Preview new dataset
print(df.head(10))
```
